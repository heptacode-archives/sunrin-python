### 1) 그림의 구조

matplotlib가 그리는 그림은 Figure 객체, Axes 객체, Axis 객체 등으로 구성된다.
Figure 객체는 한 개 이상의  Axes 객체를 포함하고 Axes 객체는 다시 두 개 이상의 Axis 객체를 포함한다.
다음 그림은 이 구조를 설명하고 있다.

### 2) 선 그래프(line plot)

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family = "Malgun Gothic")
plt.figure(dpi=300)
# color r g b k y
# marker . , ^ s o + *
plt.plot([3, 4, 5, 6, 7, 8], [13, 16, 21, 25, 29, 33], 'ro-') # x, y축 지정하고 꺾은 선 그래프 그리기
plt.axis([2.5, 8.5, 10, 40]) # 범위 지정 : x1, x2, y1, y2
plt.xlabel("월")
plt.ylabel("온도")
plt.show() # 그래프 보여주기
```

```python
import matplotlib.pyplot as plt
x = range(0, 100)
y = [v*v for v in x]
plt.plot(x, y)
plt.show()
```

```python
# y=x, y=x**2, yx**3
import matplotlib.pyplot as plt
import numpy as np
x_range = np.arange(0, 5, 0.2)
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (3,5), dpi = 150)
plt.plot(x_range, x_range)
plt.plot(x_range, x_range**2)
plt.plot(x_range, x_range**3)
plt.legend(loc = 'upper right') # upper lower right loft center
plt.axis([-0.5, 5.5, -0.5, 5.5])
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('여러 종류 그래프')
plt.savefig('여러종류그래프.png')
plt.show()
```
