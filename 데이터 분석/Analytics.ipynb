```python
import csv
data = csv.reader(open('datas/1_월별.CSV', encoding = 'cp949'))
next(data)
result = []

for row in data:
    for i in range(2, len(row)):
        tmp = ''.join(row[i].split(','))
        result.append(float(tmp))
        #result.append(row[i])


print("result :", result)

import matplotlib.pyplot as plt
plt.rc('font', family = 'Junicode-Bold')
plt.plot(result)
plt.title('서울시 한국여행중 서울지역 방문지 월별 통계 - 2018')
plt.show()
```
